```python
fl1 = pd.read_csv("../data/flights/flights2017_2019.csv", encoding = "latin1")
fl2 = pd.read_csv("../data/flights/flights2020_2022.csv", encoding = "latin1")
fl3 = pd.read_csv("../data/flights/flights.csv", encoding = "latin1")
```

```python
columnas_excepto_una = fl.columns.difference(['flight_id'])  
fl = fl.drop_duplicates(subset=columnas_excepto_una, keep='first')
fl.shape
```
